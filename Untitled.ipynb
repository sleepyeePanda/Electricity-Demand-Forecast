```python
import pandas as pd
```

```python
power_peak_df = pd.read_csv("average_monthly_power_peak.csv")
power_peak_df.head()
```

```python
power_peak_df.replace(0, pd.np.nan, inplace=True)
power_peak_df.dropna(axis=0, inplace=True)
power_peak_df
```

```python
power_peak_df = pd.melt(power_peak_df, id_vars=['연도'], var_name='월')
```

```python
power_peak_df.dtypes
```

```python
power_peak_df['date']= pd.to_datetime(power_peak_df['연도'].astype(str) + '/' + power_peak_df['월'].astype(str) + '/01')
```

```python
power_peak_df.drop(['연도', '월'], axis=1, inplace=True)
```

```python
power_peak_df.set_index('date', inplace=True)
print(power_peak_df)
```

```python

import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(power_peak_df, lags=20)
plot_pacf(power_peak_df, lags=20)
plt.show()
```
